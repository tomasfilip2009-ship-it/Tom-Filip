```python
class Zoo:
  def __init__(self, number, species, food):
    self.number = number
    self.species = species
    self.food = food
  def describe(self):
        print('There are', self.number, self.species, 'and they eat', self.food, '.')

my_zoo = Zoo(3, 'monkeys', 'Bananas')
my_zoo.describe()
```

```python
class Animal:
  def __init__(self, species, food):
    self.species = species
    self.food = food

  def describe(self):
    print(f'This is a {self.species} and it eats {self.food}.')

  def make_sound(self):
    print(f'The {self.species} makes a sound.')


class Monkey(Animal):
  def describe(self):
    super().describe()
    print('Ooh ooh aah aah!')

  def make_sound(self):
    print('Ooh ooh aah aah!')


class Lion(Animal):
  def describe(self):
    super().describe()
    print('Roar!')

  def make_sound(self):
    print('Roar!')


monkey_instance = Monkey('monkey', 'bananas')
lion_instance = Lion('lion', 'meat')

monkey_instance.describe()
monkey_instance.make_sound()

lion_instance.describe()
lion_instance.make_sound()
```
