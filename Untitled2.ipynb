```python
from ColabTurtle.Turtle import *
import math

initializeTurtle()
speed(10) # Adjusted speed for smoother drawing

# Define common parameters for both shapes
main_center_x, main_center_y = 400, 250 # Base center point for positioning
circum_radius = 90 # This will be the circumradius for the pentagram and the circle

# Adjust pentagram's vertical center slightly upwards for visual alignment
pentagram_center_y = main_center_y - 10 # Shift pentagram up by 10 pixels

# Draw the circle
pencolor('red')
width(2)
penup()
# The circle should circumscribe the pentagram, so its radius and center
# should match the pentagram's circumradius and center.
circle_radius = circum_radius
circle_center_x = main_center_x
circle_center_y = pentagram_center_y # Use the adjusted y-center for the circle

# Go to the starting point for the circle (bottom-most point)
goto(circle_center_x, circle_center_y - circle_radius)
pendown()
setheading(0) # Point east to start drawing the circle
for _ in range(360):
    forward(math.pi * 2 * circle_radius / 360)
    left(1)

# Draw the pentagram
penup()
pencolor('blue')
width(3)

vertices = []
for i in range(5):
    angle = i * 2 * math.pi / 5 - math.pi / 2
    x = main_center_x + circum_radius * math.cos(angle)
    y = pentagram_center_y + circum_radius * math.sin(angle)
    vertices.append((x, y))

goto(vertices[0][0], vertices[0][1])
pendown()
goto(vertices[2][0], vertices[2][1])
goto(vertices[4][0], vertices[4][1])
goto(vertices[1][0], vertices[1][1])
goto(vertices[3][0], vertices[3][1])
goto(vertices[0][0], vertices[0][1])
```

```python
from ColabTurtle.Turtle import *
import math
import random

initializeTurtle()
bgcolor('black')
speed(13) # Set speed for smoother animation
hideturtle()
pencolor('white') # Default pen color, will change in loop

# Define the center of the black hole
center_x, center_y = window_width() / 2, window_height() / 2

# List of dark colors for the black hole effect
dark_colors = [
    '#1a1a1a', '#333333', '#4d4d4d', '#666666', '#808080',
    '#0a0a0a', '#141414', '#1f1f1f', '#292929', '#3d3d3d',
    '#00001a', '#000033', '#00004d', '#000066', '#000080',
    '#1a001a', '#330033', '#4d004d', '#660066', '#800080'
]

max_radius = min(center_x, center_y) * 0.9
num_circles = 80 # Number of concentric circles

for i in range(num_circles):
    # Calculate radius, increasing outwards
    radius = (i / num_circles) * max_radius

    # Vary line width: wider for inner circles, thinner for outer
    # Or wider for outer to represent event horizon (adjust as desired)
    # Here, I'm making it wider for outer circles for a more dramatic effect towards the edge
    line_width = int(1 + (i / num_circles) * 5) # Cast to int
    width(line_width)

    # Choose a dark color, potentially mixing with randomness
    color_index = i % len(dark_colors)
    pen_color = dark_colors[color_index]

    # Add some random distortion to the center point for each circle
    # This creates a wobbly, distorted effect
    distortion_x = random.uniform(-5, 5)
    distortion_y = random.uniform(-5, 5)

    current_center_x = center_x + distortion_x
    current_center_y = center_y + distortion_y

    pencolor(pen_color)

    penup()
    # Move to the starting point of the circle (bottom-most point relative to its center)
    goto(current_center_x, current_center_y - radius)
    pendown()

    setheading(0) # Point east to start drawing the circle

    # Draw the circle by moving forward and turning slightly
    # A small 'wiggle' can be added for distortion within the circle itself
    circle_segments = 100 # More segments for smoother circles
    for _ in range(circle_segments):
        segment_length = math.pi * 2 * radius / circle_segments
        # Add a very slight random variation to the segment length for subtle distortion
        forward(segment_length + random.uniform(-0.1, 0.1))
        left(360 / circle_segments)

# Finalizing drawing
penup()
goto(center_x, center_y) # Move turtle to center after drawing, though hidden
print("Black hole simulation drawn.")
```

```python
from ColabTurtle.Turtle import *
import math
import random

initializeTurtle()
bgcolor('black')
speed(13)
hideturtle()

center_x, center_y = window_width() / 2, window_height() / 2

bright_colors = [
    '#ffffff', '#ffff66', '#ffcc33', '#ff99ff', '#66ccff', '#ccffcc',
    '#ff6666', '#ffccff', '#99ffff', '#ffcc99'
]

num_shapes = 40  # sneno z 120
max_radius = min(center_x, center_y) * 0.7  # men rozsah

for i in range(num_shapes):
    radius = random.uniform(0, max_radius)
    angle = random.uniform(0, 360)
    offset_x = radius * math.cos(math.radians(angle))
    offset_y = radius * math.sin(math.radians(angle))
    x, y = center_x + offset_x, center_y + offset_y

    shape_type = random.choice(['circle', 'triangle', 'star', 'polygon'])
    size = random.uniform(20, 50)  # sneno max, aby to nebylo tak velk
    pen_color = random.choice(bright_colors)
    width(random.randint(1, 2))  # men tlouka pro leh vzhled

    penup()
    goto(x, y)
    pendown()
    pencolor(pen_color)

    if shape_type == 'circle':
        penup()
        goto(x, y - size/2)
        pendown()
        circle_segments = max(20, int(size / 2))
        for _ in range(circle_segments):
            seg_len = math.pi * size / circle_segments
            forward(seg_len + random.uniform(-0.2,0.2))
            left(360 / circle_segments)
    elif shape_type == 'triangle':
        deform = size * 0.2
        for _ in range(3):
            forward(size + random.uniform(-deform,deform))
            left(120 + random.uniform(-deform,deform))
    elif shape_type == 'star':
        deform = size * 0.1
        for _ in range(5):
            forward(size)
            right(144 + random.uniform(-deform,deform))
    elif shape_type == 'polygon':
        sides = random.randint(4,6)  # mn stran pro rychlej kreslen
        deform = size * 0.15
        for _ in range(sides):
            forward(size + random.uniform(-deform,deform))
            left(360 / sides + random.uniform(-deform,deform))

penup()
goto(center_x, center_y)
print("Chaotick efekt nakreslen.")


```

```python
from ColabTurtle.Turtle import *
import math # Import the math module

# Define a circle function
def circle(radius):
    circumference = 2 * math.pi * radius
    segments = 60 # Number of segments for smoothness
    segment_length = circumference / segments
    for _ in range(segments):
        forward(segment_length)
        left(360 / segments)

# inicializace
initializeTurtle(initial_speed=13)
bgcolor("black")
hideturtle()
speed(13)

colors = [
    "#FF0000",
    "#FF7F00",
    "#FFFF00",
    "#00FF00",
    "#0000FF",
    "#4B0082",
    "#8B00FF"
]

for i in range(1, 250):
    pencolor(colors[i % len(colors)])  # barvu nastavujeme funkc pencolor()
    width((i % 5) + 1)

    forward(i * 0.8)
    left(35)
    circle(i * 0.15) # Now the custom circle function will be called
```
